```python
# ================================================================
# CRNN + MFCC + HEAVY AUGMENT (All files used, balanced-by-augment)
# ================================================================
# - Folder dataset: <DATA_DIR>/<class_name>/*.wav|*.mp3|...
# - TRAIN: balanced via augmentation (upsample kelas kecil, tanpa buang file)
# - VAL  : 1 segmen RMS tertinggi per file (deterministik)
# - Fitur: MFCC + delta + delta2  (shape: [T, 120, 1]), CRNN (Conv + BiLSTM)
# - Output: history plots + confusion matrix + TTA util
# ================================================================

import os, sys, random, warnings, importlib, subprocess
warnings.filterwarnings("ignore")

# -- optional: auto install
try:
    import audiomentations  # noqa
except Exception:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "audiomentations>=0.35.0"])

import numpy as np
from pathlib import Path
import librosa, matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers, regularizers
from audiomentations import Compose, AddGaussianNoise, TimeStretch, PitchShift, Shift, ClippingDistortion, Gain

# --------------------- Config yang bisa kamu ubah ---------------------
DATA_DIR   = r"D:\web\cnn_clasification\archive\Data\genres_original"  # ganti ke folder-mu
SEED       = 42

# Audio & segmentasi
SR         = 16000
WIN_SEC    = 2.0          # durasi segmen (disarankan 1.5 - 2.0 s)
STEP_SEC   = 0.5          # hop saat scanning RMS (untuk pilih top-K)
TOPK_TRAIN = 5            # TRAIN: sampling dari K segmen RMS tertinggi per file

# MFCC features
N_MFCC     = 40
N_FFT      = 512
HOP_LEN    = 160          # 10 ms @ 16 kHz
ADD_DELTAS = True         # True -> [mfcc, delta, delta2] = 120 fitur

# Target balancing TRAIN per kelas (upsample via augment).
# Karena datamu max ~80 per kelas, pakai 80 (kalau ada yang di bawah, akan di-augment naik).
TARGET_TRAIN_PER_CLASS = 80

# Training
EPOCHS     = 100
BATCH_SIZE = 32
LR         = 3e-4
L2W        = 1e-4
DROPOUT    = 0.35
LABEL_SMOOTH = 0.0  # set >0 jika pakai tf-keras 2.14+ yang support label_smoothing di SparseCE

np.random.seed(SEED); random.seed(SEED); tf.random.set_seed(SEED)


# --------------------- Util dasar ---------------------
def list_audio_files(root, exts=(".wav",".mp3",".flac",".ogg",".m4a",".WAV",".MP3",".FLAC")):
    root = Path(root)
    return [p for p in sorted(root.rglob("*")) if p.suffix in exts and p.is_file()]

def load_full(path, sr=SR):
    y, _ = librosa.load(path, sr=sr, mono=True)
    if np.max(np.abs(y)) > 0:
        y = y / np.max(np.abs(y))
    # pre-emphasis ringan
    y = np.append(y[0], y[1:] - 0.97 * y[:-1])
    return y

def scan_rms(y, sr=SR, win_sec=WIN_SEC, step_sec=STEP_SEC):
    win  = int(win_sec * sr)
    step = int(step_sec * sr)
    if len(y) < win:
        y = np.pad(y, (0, win - len(y)))
    starts = np.arange(0, len(y) - win + 1, step)
    if starts.size == 0:
        starts = np.array([0])
    rms = np.array([np.sqrt(np.mean(y[s:s+win]**2)) + 1e-9 for s in starts], dtype=np.float32)
    return starts, rms, win

def crop_at(y, start, win):
    end = start + win
    if end > len(y):
        y = np.pad(y, (0, end - len(y)))
    return y[start:end]

def l2_norm(seg):
    rms = np.sqrt(np.mean(seg**2)) + 1e-9
    return seg / rms


# --------------------- Augmentasi audio ---------------------
import numpy as np
from audiomentations import (
    Compose, AddGaussianNoise, TimeStretch, PitchShift,
    ClippingDistortion, Gain
)
import inspect

# --- Helper: pilih argumen Gain yang sesuai versi audiomentations ---
def make_gain(min_db=-8, max_db=6, p=0.5):
    sig = inspect.signature(Gain.__init__)
    params = sig.parameters
    # Cek nama argumen yang tersedia
    if "min_gain_in_db" in params and "max_gain_in_db" in params:
        return Gain(min_gain_in_db=min_db, max_gain_in_db=max_db, p=p)
    elif "min_gain_db" in params and "max_gain_db" in params:
        return Gain(min_gain_db=min_db, max_gain_db=max_db, p=p)
    else:
        # fallback terakhir: instansiasi tanpa bounds (kalau versi sangat lama)
        return Gain(p=p)

# --- Augment inti (tanpa Shift bawaan; kita pakai shift manual di bawah) ---
AUG = Compose([
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.01, p=0.5),
    TimeStretch(min_rate=0.85, max_rate=1.20, p=0.5),
    PitchShift(min_semitones=-3, max_semitones=3, p=0.5),
    # ClippingDistortion: versi terbaru pakai *_percentile_threshold
    ClippingDistortion(
        min_percentile_threshold=0.0,
        max_percentile_threshold=5.0,
        p=0.30
    ),
    make_gain(-8, 6, p=0.5),
])

# --- Shift manual, bebas dari perubahan API ---
def _random_time_shift(samples: np.ndarray, max_fraction: float = 0.30, p: float = 0.5):
    if np.random.rand() > p or samples.size == 0:
        return samples
    n = len(samples)
    max_shift = int(max_fraction * n)
    if max_shift < 1:
        return samples
    k = np.random.randint(-max_shift, max_shift + 1)
    if k == 0:
        return samples
    # 50% zero-pad, 50% cyclic roll
    if np.random.rand() < 0.5:
        out = np.zeros_like(samples)
        if k > 0:
            out[k:] = samples[:n - k]
        else:
            k = -k
            out[:n - k] = samples[k:]
        return out.astype(np.float32)
    else:
        return np.roll(samples, k).astype(np.float32)

def maybe_augment(seg: np.ndarray, sr: int):
    seg = AUG(samples=seg, sample_rate=sr).astype(np.float32)
    seg = _random_time_shift(seg, max_fraction=0.30, p=0.5)
    return seg




# --------------------- Ekstraksi MFCC ---------------------
def mfcc_features(seg, sr=SR, n_mfcc=N_MFCC, n_fft=N_FFT, hop=HOP_LEN, add_deltas=ADD_DELTAS):
    # MFCC [n_mfcc, T]
    mfcc = librosa.feature.mfcc(y=seg, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop)
    # normalisasi per-koef (stabilkan)
    mfcc = (mfcc - np.mean(mfcc, axis=1, keepdims=True)) / (np.std(mfcc, axis=1, keepdims=True) + 1e-8)

    feats = [mfcc]
    if add_deltas:
        d1 = librosa.feature.delta(mfcc, order=1)
        d2 = librosa.feature.delta(mfcc, order=2)
        feats.extend([d1, d2])
    feat = np.concatenate(feats, axis=0)  # [F, T]; F = 40 / 80 / 120

    # pad/truncate ke panjang waktu tetap (berdasar WIN_SEC & hop)
    T_target = int(np.ceil((len(seg) / hop)))   # kurang lebih
    T_fix = int(np.ceil(WIN_SEC * sr / hop))
    if T_target < T_fix:
        pad = np.zeros((feat.shape[0], T_fix - feat.shape[1]), dtype=np.float32)
        feat = np.concatenate([feat, pad], axis=1)
    else:
        feat = feat[:, :T_fix]

    # [T, F, 1]
    feat = np.transpose(feat, (1, 0))[:, :, None]
    return feat.astype(np.float32)


# --------------------- Build dataset ---------------------
def build_dataset(data_dir, mode="train", target_per_class=TARGET_TRAIN_PER_CLASS, topk_train=TOPK_TRAIN):
    """
    mode='train': upsample via augment hingga target_per_class (tanpa buang file)
    mode='val'  : 1 segmen RMS tertinggi per file (deterministik, no augment)
    Output: X:[N, T, F, 1], y:[N], class_names
    """
    root = Path(data_dir)
    class_names = sorted([d.name for d in root.iterdir() if d.is_dir()])
    X_list, y_list, paths = [], [], []

    for ci, cname in enumerate(class_names):
        files = list_audio_files(root / cname)
        if mode == "train":
            target = max(target_per_class, len(files))   # pakai semua file, kecil di-augment naik
            rep = int(np.ceil(target / max(1, len(files))))
        else:
            target = len(files)
            rep = 1
        print(f"[{mode.upper()}] {cname}: {len(files)} file -> {rep} crop/file (target≈{target})")

        for f in files:
            try:
                yfull = load_full(str(f), sr=SR)
                starts, rms, win = scan_rms(yfull, sr=SR)

                if mode == "train":
                    pick = np.argsort(rms)[::-1][:min(topk_train, len(rms))]
                    for _ in range(rep):
                        s = int(np.random.choice(pick))
                        seg = crop_at(yfull, starts[s], win)
                        seg = l2_norm(seg)
                        seg = maybe_augment(seg, sr=SR)
                        feat = mfcc_features(seg)
                        X_list.append(feat); y_list.append(ci); paths.append(str(f))
                else:
                    s = int(np.argmax(rms))
                    seg = crop_at(yfull, starts[s], win)
                    seg = l2_norm(seg)
                    feat = mfcc_features(seg)
                    X_list.append(feat); y_list.append(ci); paths.append(str(f))
            except Exception as e:
                print("  Gagal:", f, e)

    X = np.stack(X_list, axis=0)
    y = np.array(y_list, dtype=np.int64)
    return X, y, class_names, np.array(paths)


print("==> Build TRAIN ...")
X_train, y_train, class_names, _ = build_dataset(DATA_DIR, mode="train")
print("==> Build VAL ...")
X_val,   y_val,   _, _          = build_dataset(DATA_DIR, mode="val")

print("Shapes:", X_train.shape, X_val.shape, "| classes:", len(class_names), class_names)


# --------------------- Model: CRNN ---------------------
from tensorflow.keras import layers, models, regularizers

def conv_block(x, filters, k=(3,3), p=(2,2), drop=0.3, l2w=2e-4):
    x = layers.Conv2D(filters, k, padding="same",
                      kernel_regularizer=regularizers.l2(l2w))(x)
    x = layers.BatchNormalization()(x); x = layers.Activation("relu")(x)
    x = layers.Conv2D(filters, k, padding="same",
                      kernel_regularizer=regularizers.l2(l2w))(x)
    x = layers.BatchNormalization()(x); x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D(p)(x)
    x = layers.Dropout(drop)(x)
    return x

def build_crnn(input_shape, num_classes, l2w=2e-4, drop=0.35, rnn_units=128):
    """
    input_shape: (T, F, C) -> mis. (time, mel_bins, 1)
      - Pastikan F (mel bins) dan C diketahui (INTEGER), T boleh variabel.
    """
    inp = layers.Input(shape=input_shape)          # [B, T, F, C]

    # --- CNN feature extractor ---
    x = conv_block(inp, 32,  drop=0.30)
    x = conv_block(x,   64,  drop=0.35)
    x = conv_block(x,   96,  drop=0.35)
    # x: [B, T', F', C]

    # --- Ratakan dimensi frekuensi & channel per time step (STATIC) ---
    # Ini memastikan dimensi fitur untuk GRU bukan None.
    x = layers.TimeDistributed(layers.Flatten(), name="td_flatten")(x)  # [B, T', F'*C]

    # --- BiGRU head ---
    x = layers.Bidirectional(layers.GRU(rnn_units, return_sequences=True))(x)
    x = layers.Bidirectional(layers.GRU(rnn_units))(x)

    # --- FC classifier ---
    x = layers.Dense(256, activation="relu", kernel_regularizer=regularizers.l2(l2w))(x)
    x = layers.Dropout(drop)(x)
    out = layers.Dense(num_classes, activation="softmax")(x)

    return models.Model(inp, out, name="CRNN")



input_shape = X_train.shape[1:]   # [T, F, 1]
num_classes = len(class_names)
model = build_crnn(input_shape, num_classes)
model.summary()

# loss (fallback untuk tf-keras lawas tanpa label_smoothing param)
try:
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(label_smoothing=LABEL_SMOOTH)
except TypeError:
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()

model.compile(
    optimizer=optimizers.Adam(LR),
    loss=loss_fn,
    metrics=["accuracy"]
)

cbs = [
    callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6, verbose=1),
    callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True, verbose=1),
    callbacks.ModelCheckpoint("best_crnn_mfcc.keras", monitor="val_loss", save_best_only=True, verbose=1),
]

hist = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=cbs,
    verbose=1
)

# --------------------- Evaluasi ---------------------
y_prob = model.predict(X_val, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("Val acc:", accuracy_score(y_val, y_pred))
print(classification_report(y_val, y_pred, target_names=class_names, digits=4))
cm = confusion_matrix(y_val, y_pred)
print("Confusion:\n", cm)

# --------------------- Plot helper ---------------------
def plot_history(h):
    plt.figure(figsize=(7,5))
    plt.plot(h.history['accuracy'], label='train acc')
    plt.plot(h.history['val_accuracy'], label='val acc')
    plt.title('Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Acc')
    plt.grid(True, alpha=.3); plt.legend(); plt.tight_layout(); plt.show()

    plt.figure(figsize=(7,5))
    plt.plot(h.history['loss'], label='train loss')
    plt.plot(h.history['val_loss'], label='val loss')
    plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss')
    plt.grid(True, alpha=.3); plt.legend(); plt.tight_layout(); plt.show()

def plot_cm(cm, class_names, normalize=False, title="Confusion Matrix"):
    data = cm.astype(float)
    fmt = "d"
    if normalize:
        row_sums = data.sum(axis=1, keepdims=True) + 1e-9
        data = data / row_sums
        fmt = ".2f"
    plt.figure(figsize=(7,6))
    plt.imshow(data, interpolation="nearest", cmap=plt.cm.Blues)
    plt.title(title); plt.colorbar(fraction=0.046, pad=0.04)
    ticks = np.arange(len(class_names))
    plt.xticks(ticks, class_names, rotation=45, ha="right")
    plt.yticks(ticks, class_names)
    thresh = data.max() / 2.0
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            plt.text(j, i, format(data[i, j], fmt),
                     ha="center", va="center",
                     color="white" if data[i, j] > thresh else "black")
    plt.ylabel("True"); plt.xlabel("Predicted"); plt.tight_layout(); plt.show()

plot_history(hist)
plot_cm(cm, class_names, normalize=False, title="Confusion Matrix (VAL)")
plot_cm(cm, class_names, normalize=True,  title="Normalized Confusion Matrix (VAL)")


# --------------------- TTA (multi-crop voting) untuk 1 file ---------------------
def predict_file_tta(file_path, model, starts_sec=(0.0, 0.25, 0.5, 0.75, 1.0),
                     sr=SR, win_sec=WIN_SEC):
    y = load_full(file_path, sr=sr)
    win = int(win_sec * sr)
    probs = []
    for st in starts_sec:
        s = int(st * sr); e = s + win
        if e > len(y):
            seg = np.pad(y[s:], (0, e - len(y)))
        else:
            seg = y[s:e]
        seg = l2_norm(seg)
        feat = mfcc_features(seg)  # [T, F, 1]
        p = model.predict(feat[None, ...], verbose=0)[0]
        probs.append(p)
    p_mean = np.mean(probs, axis=0)
    return p_mean, int(np.argmax(p_mean))

# contoh pakai:
# p, cls = predict_file_tta(r"D:\path\to\audio.wav", model)
# print("Pred:", class_names[cls], "probs:", p)
```
