```python
# =========================
# ADVANCED AUDIO CLASSIFICATION - COMPLETE (ML + DNN + CNN)
# =========================
import os
import numpy as np
import librosa
from pathlib import Path
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Library Machine Learning Klasik
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Library Deep Learning (TensorFlow/Keras)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv1D, MaxPooling1D, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

warnings.filterwarnings("ignore")

# ---------- Konfigurasi ----------
# Ganti path ini sesuai lokasi dataset Anda
DATA_DIR = r"D:\web\cnn_clasification\archive\Data\genres_original"
SR = 22050
DURATION = 2.0 
SEED = 42

np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---------- 1. ADVANCED FEATURE EXTRACTION ----------
def extract_advanced_features(audio_path, sr=22050, duration=2.0):
    """Ekstraksi fitur audio komprehensif untuk input model"""
    try:
        y, sr = librosa.load(audio_path, sr=sr, duration=duration, mono=True)
        # Trim silence
        y_trimmed, _ = librosa.effects.trim(y, top_db=25)
        if len(y_trimmed) > 0: y = y_trimmed
        
        # Padding/Truncating ke panjang tetap
        target_length = int(sr * duration)
        if len(y) < target_length:
            y = np.pad(y, (0, target_length - len(y)), mode='constant')
        else:
            y = y[:target_length]
        
        features = []
        
        # --- MFCC & Delta ---
        mfcc13 = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, hop_length=512)
        mfcc20 = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20, hop_length=512)
        mfcc_delta = librosa.feature.delta(mfcc13)
        mfcc_delta2 = librosa.feature.delta(mfcc13, order=2)
        
        for feat in [mfcc13, mfcc20, mfcc_delta, mfcc_delta2]:
            features.extend(np.mean(feat, axis=1))
            features.extend(np.std(feat, axis=1))
        
        # --- Spectral Features ---
        spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
        spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
        
        for feat in [spectral_centroid, spectral_rolloff, spectral_bandwidth, contrast]:
            features.extend(np.mean(feat, axis=1)) # Note: contrast returns multiple bands
            features.extend(np.std(feat, axis=1))
            
        # --- Chroma & Tonnetz ---
        chroma = librosa.feature.chroma_stft(y=y, sr=sr)
        tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
        
        for feat in [chroma, tonnetz]:
            features.extend(np.mean(feat, axis=1))
            features.extend(np.std(feat, axis=1))
        
        # --- Rhythm & Energy ---
        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        features.append(tempo)
        
        zcr = librosa.feature.zero_crossing_rate(y)
        features.append(np.mean(zcr))
        features.append(np.std(zcr))
        
        rms = librosa.feature.rms(y=y)
        features.append(np.mean(rms))
        features.append(np.std(rms))
        
        # --- Mel Spectrogram Stats ---
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=64)
        mel_db = librosa.power_to_db(mel)
        features.append(np.mean(mel_db))
        features.append(np.std(mel_db))
        features.append(np.max(mel_db))
        features.append(np.min(mel_db))
        
        return np.nan_to_num(np.array(features, dtype=np.float32))
        
    except Exception as e:
        print(f"❌ Error processing {Path(audio_path).name}: {e}")
        return None

# ---------- 2. DATA COLLECTION ----------
def collect_advanced_dataset(data_dir, samples_per_class=20):
    """Mengumpulkan data audio dari folder"""
    data_root = Path(data_dir)
    class_dirs = [d for d in data_root.iterdir() if d.is_dir()]
    class_dirs.sort()
    class_names = [d.name for d in class_dirs]
    
    print(f"🎵 COLLECTING DATASET: {len(class_names)} Classes found")
    X, y = [], []
    
    for class_idx, class_dir in enumerate(class_dirs):
        # Cari file wav dan mp3
        audio_files = list(class_dir.glob("*.wav")) + list(class_dir.glob("*.mp3"))
        
        # Limit sample jika diminta
        if len(audio_files) > samples_per_class: 
            audio_files = audio_files[:samples_per_class]
        
        print(f"   📂 Processing {class_names[class_idx]} ({len(audio_files)} files)...")
        
        processed_count = 0
        for audio_file in audio_files:
            features = extract_advanced_features(str(audio_file))
            if features is not None:
                X.append(features)
                y.append(class_idx)
                processed_count += 1
        
        if processed_count == 0:
            print(f"      ⚠️ No valid features extracted for {class_names[class_idx]}")
    
    # Fallback ke data sintetik jika tidak ada file audio ditemukan (misal path salah)
    if len(X) == 0: 
        print("❌ No audio data found! Switching to SYNTHETIC data for testing...")
        return create_advanced_synthetic_data(class_names, samples_per_class)
        
    return np.array(X), np.array(y), class_names

def create_advanced_synthetic_data(class_names, samples_per_class=20):
    """Membuat data dummy jika file asli tidak ditemukan"""
    print("🔧 GENERATING SYNTHETIC DATA")
    n_classes = len(class_names)
    if n_classes == 0:
        class_names = ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']
        n_classes = len(class_names)
        
    n_features = 158 # Estimasi jumlah fitur dari ekstraksi
    X, y = [], []
    for class_idx in range(n_classes):
        for _ in range(samples_per_class):
            # Buat pola data berbeda tiap kelas
            features = np.random.normal(class_idx * 0.5, 1.0, n_features)
            # Tambah noise
            features += np.random.normal(0, 0.2, n_features)
            X.append(features)
            y.append(class_idx)
    return np.array(X), np.array(y), class_names

# ---------- 3. VISUALIZATION FUNCTIONS ----------
def plot_history(history, model_name):
    """Plot Accuracy dan Loss dari training history"""
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(14, 5))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'b-', label='Training Acc')
    plt.plot(epochs, val_acc, 'r-', label='Validation Acc')
    plt.title(f'{model_name} - Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title(f'{model_name} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(model, X_test, y_test, class_names, model_name):
    """Plot Confusion Matrix untuk model"""
    # Prediksi
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    
    # Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# ---------- 4. TRAINING DEEP LEARNING (DNN & CNN) ----------
def train_dl_models(X, y, class_names):
    print("\n🧠 TRAINING DEEP LEARNING MODELS (DNN & CNN)")
    print("=" * 50)

    # Split Data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=SEED, stratify=y)
    
    # Scaling (Wajib untuk NN)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    input_shape = X_train.shape[1]
    num_classes = len(class_names)
    epochs = 50
    batch_size = 16

    # Callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001)

    results = {}

    # --- A. DNN (Deep Neural Network) ---
    print("\n🏗️ Building DNN Model...")
    dnn_model = Sequential([
        Dense(512, input_shape=(input_shape,), activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])

    dnn_model.compile(loss='sparse_categorical_crossentropy', 
                      optimizer=Adam(learning_rate=0.001), 
                      metrics=['accuracy'])

    print("🚀 Training DNN...")
    history_dnn = dnn_model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )
    
    # Visualisasi DNN (Dipanggil DI DALAM fungsi agar variabel terlihat)
    print("📊 Visualizing DNN Results...")
    plot_history(history_dnn, "DNN Model")
    plot_confusion_matrix(dnn_model, X_test, y_test, class_names, "DNN Model")
    
    loss, acc = dnn_model.evaluate(X_test, y_test, verbose=0)
    results['DNN'] = acc
    print(f"✅ DNN Test Accuracy: {acc:.4f}")

    # --- B. CNN 1D (Convolutional Neural Network) ---
    print("\n🏗️ Building CNN 1D Model...")
    
    # Reshape untuk CNN: (samples, features, 1)
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    cnn_model = Sequential([
        Conv1D(64, kernel_size=3, padding='same', activation='relu', input_shape=(input_shape, 1)),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Conv1D(128, kernel_size=3, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Conv1D(256, kernel_size=3, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.3),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.4),
        Dense(num_classes, activation='softmax')
    ])

    cnn_model.compile(loss='sparse_categorical_crossentropy', 
                      optimizer=Adam(learning_rate=0.001), 
                      metrics=['accuracy'])

    print("🚀 Training CNN...")
    history_cnn = cnn_model.fit(
        X_train_cnn, y_train,
        validation_data=(X_test_cnn, y_test),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )

    # Visualisasi CNN
    print("📊 Visualizing CNN Results...")
    plot_history(history_cnn, "CNN 1D Model")
    plot_confusion_matrix(cnn_model, X_test_cnn, y_test, class_names, "CNN 1D Model")

    loss, acc = cnn_model.evaluate(X_test_cnn, y_test, verbose=0)
    results['CNN'] = acc
    print(f"✅ CNN Test Accuracy: {acc:.4f}")

    return results

# ---------- 5. TRAINING CLASSIC ML ----------
def train_classic_models(X, y, class_names):
    print("\n🤖 TRAINING CLASSIC MODELS")
    print("=" * 50)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=SEED, stratify=y)
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    models = {
        'Random Forest': RandomForestClassifier(n_estimators=100, random_state=SEED),
        'SVM RBF': SVC(kernel='rbf', random_state=SEED),
    }
    
    results = {}
    for name, model in models.items():
        print(f"⚙️ Training {name}...")
        model.fit(X_train_scaled, y_train)
        acc = model.score(X_test_scaled, y_test)
        results[name] = acc
        print(f"   ✅ {name} Accuracy: {acc:.4f}")
        
        # --- BAGIAN INI SUDAH SAYA UN-COMMENT AGAR MUNCUL ---
        print(f"   📊 Plotting Confusion Matrix for {name}...")
        y_pred = model.predict(X_test_scaled)
        cm = confusion_matrix(y_test, y_pred)
        
        plt.figure(figsize=(8,6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
        plt.title(f'Confusion Matrix - {name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.tight_layout()
        plt.show()
        # ----------------------------------------------------
    
    return results

# ---------- MAIN EXECUTION ----------
def main():
    print("=== 🎵 ADVANCED AUDIO CLASSIFICATION SYSTEM ===")
    print("=== Features: MFCC, Spectral, Rhythm, Tonnetz ===")
    print("=== Models: SVM, Random Forest, DNN, CNN 1D ===")
    print("=" * 60)
    
    try:
        # 1. Collect Data
        X, y, class_names = collect_advanced_dataset(DATA_DIR, samples_per_class=50)
        
        # 2. Train Classic Models
        classic_results = train_classic_models(X, y, class_names)
        
        # 3. Train Deep Learning Models
        dl_results = train_dl_models(X, y, class_names)
        
        # 4. Final Comparison
        print("\n" + "=" * 60)
        print("🏆 FINAL MODEL LEADERBOARD")
        print("=" * 60)
        
        all_results = {**classic_results, **dl_results}
        sorted_results = dict(sorted(all_results.items(), key=lambda item: item[1], reverse=True))
        
        # Print Text Table
        print(f"{'MODEL':<20} | {'ACCURACY':<10}")
        print("-" * 35)
        for model_name, acc in sorted_results.items():
            print(f"{model_name:<20} | {acc:.4f}")
            
        # Plot Comparison Chart
        plt.figure(figsize=(10, 6))
        bars = plt.bar(sorted_results.keys(), sorted_results.values(), color=sns.color_palette("viridis", len(sorted_results)))
        plt.title('Final Accuracy Comparison')
        plt.ylabel('Accuracy')
        plt.ylim(0, 1.1)
        plt.grid(axis='y', alpha=0.3)
        
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                     f'{height:.4f}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        print("\n✅ Process Completed Successfully!")
        
    except Exception as e:
        print(f"\n❌ CRITICAL ERROR IN MAIN: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```
