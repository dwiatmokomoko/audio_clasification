```python
import os
import numpy as np
import librosa
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

# Import Deep Learning Framework
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import to_categorical, Sequence
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# ---------- Konfigurasi Hyperparameter Baru ----------
DATA_DIR = r"D:\web\cnn_clasification\archive\Data\genres_original"
SR = 22050
DURATION = 1.5      # FOKUS PERBAIKAN: Dikurangi dari 3.0s menjadi 1.5s
N_MELS = 128
N_FTT = 2048
HOP_LENGTH = 256    # FOKUS PERBAIKAN: Dikurangi dari 512 menjadi 256 (resolusi waktu lebih halus)
SEED = 42
SAMPLES_PER_CLASS = 80 
BATCH_SIZE = 16     # Batch size dipertahankan kecil untuk stabilitas

np.random.seed(SEED)
tf.random.set_seed(SEED)

# Global variable to store normalized shape
EXPECTED_SHAPE = None

# ---------- UTILITIES FOR FEATURE EXTRACTION & AUGMENTATION ----------

def extract_mel_spectrogram(y, sr=SR, n_mels=N_MELS, n_fft=N_FTT, hop_length=HOP_LENGTH):
    """Menghitung dan menormalisasi Mel Spectrogram."""
    mel_spectrogram = librosa.feature.melspectrogram(
        y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels
    )
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
    
    # Normalisasi Min-Max
    min_val = np.min(mel_spectrogram_db)
    max_val = np.max(mel_spectrogram_db)
    
    if max_val - min_val == 0:
        normalized_spectrogram = np.zeros_like(mel_spectrogram_db)
    else:
        normalized_spectrogram = (mel_spectrogram_db - min_val) / (max_val - min_val)
    
    return normalized_spectrogram.T # Transpose (time_steps, n_mels)

def process_audio(audio_path, sr=SR, duration=DURATION, augment=False):
    """Load, trim, pad/truncate, Augment, dan Feature Extraction."""
    try:
        y, sr = librosa.load(audio_path, sr=sr, duration=duration, mono=True)
        
        # 1. Trim silence
        y_trimmed, _ = librosa.effects.trim(y, top_db=25)
        if len(y_trimmed) > 0:
            y = y_trimmed
            
        # 2. Augmentasi Data
        if augment:
            # Random Pitch Shift (-3 hingga +3 semitone)
            y = librosa.effects.pitch_shift(y, sr=sr, n_steps=np.random.uniform(-3, 3))
            # Random Time Stretch (0.85 hingga 1.15x)
            rate = np.random.uniform(0.85, 1.15)
            y = librosa.effects.time_stretch(y, rate=rate)

        # 3. Fixed length (padding/truncating)
        target_length = int(sr * duration)
        if len(y) < target_length:
            y = np.pad(y, (0, target_length - len(y)), mode='constant')
        else:
            y = y[:target_length]
        
        # 4. Feature Extraction
        spectrogram = extract_mel_spectrogram(y, sr)
        
        # Tambahkan channel dimension: (time_steps, n_mels, 1)
        return spectrogram[..., np.newaxis]
        
    except Exception as e:
        return None

# ---------- DATA GENERATOR WITH AUGMENTATION (No Change) ----------

class AudioDataGenerator(Sequence):
    """Data generator Keras dengan augmentasi on-the-fly."""

    def __init__(self, file_paths, labels, batch_size, input_shape, augment=False, shuffle=True):
        self.file_paths = file_paths
        self.labels = labels
        self.batch_size = batch_size
        self.input_shape = input_shape
        self.augment = augment
        self.shuffle = shuffle
        self.on_epoch_end()

    def __len__(self):
        return int(np.floor(len(self.file_paths) / self.batch_size))

    def __getitem__(self, index):
        indexes = self.indexes[index * self.batch_size:(index + 1) * self.batch_size]
        list_paths_batch = [self.file_paths[k] for k in indexes]
        list_labels_batch = [self.labels[k] for k in indexes]
        X, y = self.__data_generation(list_paths_batch, list_labels_batch)
        return X, y

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.file_paths))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __data_generation(self, list_paths_batch, list_labels_batch):
        X = np.empty((self.batch_size, *self.input_shape))
        y = np.empty((self.batch_size, self.labels.shape[1]), dtype=int)

        for i, (path, label) in enumerate(zip(list_paths_batch, list_labels_batch)):
            features = process_audio(path, augment=self.augment)
            
            if features is not None and features.shape == self.input_shape:
                X[i,] = features
                y[i] = label
            else:
                # Fallback untuk sampel yang gagal dimuat/diproses
                X[i,] = np.random.rand(*self.input_shape)
                y[i] = label

        return X, y

# ---------- ROBUST DATA COLLECTION (Updated for CNN) ----------
def collect_cnn_dataset(data_dir, samples_per_class=SAMPLES_PER_CLASS):
    """Kumpulkan file paths dan label, BUKAN feature array."""
    data_root = Path(data_dir)
    class_dirs = [d for d in data_root.iterdir() if d.is_dir()]
    class_dirs.sort()
    class_names = [d.name for d in class_dirs]
    
    print("🎵 COLLECTING FILE PATHS FOR CNN GENERATOR (DURATION=1.5s, HOP_LENGTH=256)")
    print(f"Classes: {class_names}")
    print("=" * 60)
    
    file_paths = []
    y_labels = [] 
    
    global EXPECTED_SHAPE
    
    for class_idx, class_dir in enumerate(class_dirs):
        audio_files = []
        for ext in [".wav", ".mp3", ".au"]:
            audio_files.extend(list(class_dir.glob(f"*{ext}")))
        
        if len(audio_files) > samples_per_class:
            audio_files = audio_files[:samples_per_class]
            
        for audio_file in audio_files:
            file_paths.append(str(audio_file))
            y_labels.append(class_names[class_idx])

        if EXPECTED_SHAPE is None and len(audio_files) > 0:
            sample_features = process_audio(str(audio_files[0]), augment=False)
            if sample_features is not None:
                EXPECTED_SHAPE = sample_features.shape
                print(f"📏 Expected Spectrogram Shape: {EXPECTED_SHAPE}")
    
    if len(file_paths) == 0 or EXPECTED_SHAPE is None:
        print("❌ No features extracted! Cannot proceed.")
        return None, None, None

    le = LabelEncoder()
    y_encoded = le.fit_transform(y_labels)
    class_names_sorted = list(le.classes_) 
    y_categorical = to_categorical(y_encoded)

    print(f"\n📊 DATASET READY FOR GENERATOR!")
    print(f"  Total samples: {len(file_paths)}")
    print(f"  Input Shape per sample: {EXPECTED_SHAPE}")
    print(f"  Classes: {len(class_names_sorted)}")
    
    return file_paths, y_categorical, class_names_sorted

# ---------- CNN MODEL BUILDING (Simplified) ----------
def create_simplified_cnn_model(input_shape, num_classes):
    """Membuat model CNN yang lebih sederhana untuk data kecil."""
    
    # Pengaturan L2 Regularization yang moderat
    l2_reg = l2(0.001) 

    model = Sequential([
        # Convolutional Block 1 (Reduced Complexity)
        Conv2D(32, (5, 5), activation='relu', input_shape=input_shape, padding='same', kernel_regularizer=l2_reg),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4), # Dropout dinaikkan
        
        # Convolutional Block 2 (Reduced Complexity)
        Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4),
        
        # Classifier Block
        Flatten(),
        Dense(128, activation='relu', kernel_regularizer=l2_reg), # Reduced from 256 to 128 neurons
        BatchNormalization(),
        Dropout(0.6), # Dropout tinggi
        Dense(num_classes, activation='softmax')
    ])
    
    # Compiler dengan Learning Rate yang lebih kecil
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
    
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# ---------- CNN MODEL TRAINING (Updated Early Stopping) ----------
def train_cnn_model(file_paths, y_categorical, class_names):
    """Train model CNN menggunakan Data Generator."""

    print("\n🤖 TRAINING SIMPLIFIED CNN DEEP LEARNING MODEL")
    print("=" * 60)
    
    X_train_paths, X_test_paths, y_train, y_test = train_test_split(
        file_paths, y_categorical, test_size=0.2, random_state=SEED, stratify=y_categorical
    )
    
    input_shape = EXPECTED_SHAPE
    num_classes = len(class_names)

    # Membuat Data Generators
    train_generator = AudioDataGenerator(
        X_train_paths, y_train, BATCH_SIZE, input_shape, augment=True, shuffle=True
    )
    test_generator = AudioDataGenerator(
        X_test_paths, y_test, BATCH_SIZE, input_shape, augment=False, shuffle=False
    )
    
    print(f"📈 Data Split:")
    print(f"  Training: {len(X_train_paths)} samples ({len(train_generator)} batches)")
    print(f"  Testing: {len(X_test_paths)} samples ({len(test_generator)} batches)")
    print(f"  Input Shape: {input_shape}")
    
    # Buat Model
    model = create_simplified_cnn_model(input_shape, num_classes)
    model.summary()
    
    # Training dengan Early Stopping yang lebih sabar
    # Patience ditingkatkan menjadi 20 untuk memberi model lebih banyak waktu untuk konvergen
    early_stop = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=20, restore_best_weights=True, mode='min'
    )
    
    history = model.fit(
        train_generator,
        validation_data=test_generator,
        epochs=200, 
        callbacks=[early_stop],
        verbose=1
    )
    
    # Evaluation
    print("\nEvaluating Model...")
    loss, accuracy = model.evaluate(test_generator, verbose=0)
    print(f"  ✅ Test Accuracy: {accuracy:.4f}")
    
    # Detailed analysis
    y_pred_probs = model.predict(test_generator, verbose=0)
    y_pred = np.argmax(y_pred_probs, axis=1)
    
    # Ambil label True dari generator
    y_true_list = []
    for i in range(len(test_generator)):
        _, batch_y = test_generator.__getitem__(i)
        y_true_list.append(batch_y)
    y_true = np.argmax(np.concatenate(y_true_list), axis=1)

    # Detailed classification report
    print(f"\n📈 DETAILED CLASSIFICATION REPORT:")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
    
    # Confusion matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - Simplified CNN Model\nAccuracy: {accuracy:.4f}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Plot Training History
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Save best model
    model_info = {
        'model': model,
        'class_names': class_names,
        'accuracy': accuracy,
        'input_shape': input_shape
    }
    
    joblib.dump(model_info, 'simplified_cnn_audio_model.pkl')
    print(f"💾 Simplified CNN model saved as 'simplified_cnn_audio_model.pkl'")
    
    return accuracy, 'Simplified CNN', None

# ---------- MAIN EXECUTION ----------
def main():
    print("=== 🎵 ADVANCED AUDIO CLASSIFICATION (SIMPLIFIED CNN) ===")
    print("=" * 60)
    
    try:
        # Step 1: Collect advanced dataset (File Paths)
        file_paths, y_categorical, class_names = collect_cnn_dataset(DATA_DIR, samples_per_class=SAMPLES_PER_CLASS)
        
        if file_paths is None:
             print("Exiting due to data collection error.")
             return 0
        
        # Step 2: Train CNN Deep Learning model
        accuracy, best_model, _ = train_cnn_model(file_paths, y_categorical, class_names)
        
        # Final analysis
        print("\n" + "=" * 60)
        print("🎉 FINAL ANALYSIS")
        print("=" * 60)
        
        if accuracy > 0.8:
            print(f"✅ OUTSTANDING! Accuracy: {accuracy:.4f}")
        elif accuracy > 0.6:
            print(f"✅ VERY GOOD! Accuracy: {accuracy:.4f}")
        elif accuracy > 0.4:
            print(f"⚠️  MODERATE: Accuracy: {accuracy:.4f}")
            print("  Model sudah membaik, mungkin perlu data lebih banyak.")
        else:
            print(f"❌ NEEDS WORK: Accuracy: {accuracy:.4f}")
            print("  Pertimbangkan beralih ke model RNN (GRU/LSTM).")
            
        print(f"\n📊 Details:")
        print(f"  Best Model: {best_model}")
        print(f"  Total Samples: {len(file_paths)}")
        print(f"  Classes: {len(class_names)}")
        
        return accuracy
        
    except Exception as e:
        print(f"\n❌ ERROR: {e}")
        import traceback
        traceback.print_exc()
        return 0

if __name__ == "__main__":
    accuracy = main()
```
