```python
# =========================
# CNN (mel-spectrogram) end-to-end
# =========================
import os, random, warnings
warnings.filterwarnings("ignore")

import numpy as np
from pathlib import Path
import librosa

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks, optimizers
import matplotlib.pyplot as plt

# --------- Konfigurasi ---------
DATA_DIR   = r"D:\web\cnn_clasification\archive\Data\genres_original"  # TODO: ganti ke foldermu
SUBDIR     = ""            # "" bila multi-kelas via subfolder; atau "audioanjing" untuk 1 folder
SR         = 22050
N_MELS     = 128
N_FFT      = 2048
HOP        = 512
FMIN, FMAX = 20, 8000
DURATION   = 5.0          # None = full file; atau set (mis. 5.0 detik) agar seragam
PCTL_MAX   = 95            # panjang waktu dipotong di persentil ini agar padding tidak terlalu panjang
SEED       = 42

np.random.seed(SEED); random.seed(SEED); tf.random.set_seed(SEED)

# --------- Util ---------
def list_audio_files(root, exts=(".wav",".mp3",".flac",".ogg",".m4a",".WAV",".MP3",".FLAC")):
    root = Path(root)
    return [p for p in sorted(root.rglob("*")) if p.suffix in exts and p.is_file()]

def load_audio(path, sr=SR, duration=DURATION):
    y, sr = librosa.load(path, sr=sr, duration=duration, mono=True)
    # normalisasi amplitudo
    if np.max(np.abs(y)) > 0:
        y = y / np.max(np.abs(y))
    return y, sr

def mel_spectrogram_3ch(y, sr, n_mels=N_MELS, n_fft=N_FFT, hop=HOP, fmin=FMIN, fmax=FMAX):
    """
    Return 3-channel: [log-mel, delta, delta^2], shape: (n_mels, T, 3)
    """
    S = librosa.feature.melspectrogram(
        y=y, sr=sr, n_fft=n_fft, hop_length=hop,
        n_mels=n_mels, fmin=fmin, fmax=fmax, power=2.0
    )  # (n_mels, T)
    logS = librosa.power_to_db(S, ref=np.max)  # (n_mels, T) ~ [-80,0]
    d1   = librosa.feature.delta(logS)
    d2   = librosa.feature.delta(logS, order=2)
    x    = np.stack([logS, d1, d2], axis=-1).astype(np.float32)  # (n_mels, T, 3)
    return x

def build_from_dir(data_dir, subdir=""):
    X, y, paths = [], [], []
    root = Path(data_dir)
    if subdir:
        class_names = [subdir]
        files = list_audio_files(root / subdir)
        print(f"Menemukan {len(files)} file di {root/subdir}")
        for f in files:
            try:
                yt, sr = load_audio(str(f))
                x = mel_spectrogram_3ch(yt, sr)
                X.append(x); y.append(0); paths.append(str(f))
            except Exception as e:
                print("Gagal:", f, e)
    else:
        class_names = sorted([d.name for d in root.iterdir() if d.is_dir()])
        print("Kelas:", class_names)
        for ci, cname in enumerate(class_names):
            files = list_audio_files(root / cname)
            print(f"[{cname}] {len(files)} file")
            for f in files:
                try:
                    yt, sr = load_audio(str(f))
                    x = mel_spectrogram_3ch(yt, sr)
                    X.append(x); y.append(ci); paths.append(str(f))
                except Exception as e:
                    print("Gagal:", f, e)
    return X, np.array(y, dtype=np.int64), class_names, np.array(paths)

# --------- Muat data (variabel panjang waktu) ---------
X_list, y, class_names, paths = build_from_dir(DATA_DIR, SUBDIR)
assert len(X_list) > 0, "Tidak ada file audio yang ditemukan. Cek DATA_DIR/SUBDIR."
num_classes = int(np.max(y)) + 1
heights = [xi.shape[0] for xi in X_list]
widths  = np.array([xi.shape[1] for xi in X_list])
channels = X_list[0].shape[2]
assert all(h == heights[0] for h in heights), "n_mels harus konsisten"
max_len = int(np.percentile(widths, PCTL_MAX))
max_len = max(32, max_len)
print(f"Input HxWxc: {N_MELS}x{max_len}x{channels} | Kelas: {num_classes} {class_names}")

# --------- Padding/Truncate & Normalisasi per-sample ---------
def pad_or_truncate_spec(x, max_len):
    H, W, C = x.shape
    if W >= max_len:
        return x[:, :max_len, :]
    out = np.zeros((H, max_len, C), dtype=np.float32)
    out[:, :W, :] = x
    # padding tetap 0 (setara ~ -80 dB setelah normalisasi skala)
    return out

def normalize_per_sample(x):
    # z-score per channel (stabil untuk model)
    # hindari div 0
    x = x.copy()
    for c in range(x.shape[-1]):
        mu = np.mean(x[..., c])
        sd = np.std(x[..., c]) + 1e-6
        x[..., c] = (x[..., c] - mu) / sd
    return x

X_pad = np.stack([pad_or_truncate_spec(x, max_len) for x in X_list], axis=0)  # (N, H, W, C)
X_pad = np.stack([normalize_per_sample(x) for x in X_pad], axis=0)
print("X_pad:", X_pad.shape, "y:", y.shape)

# --------- Split ---------
X_train, X_val, y_train, y_val = train_test_split(
    X_pad, y, test_size=0.2, stratify=y, random_state=SEED
)

# --------- Class weights ---------
class_weights_arr = compute_class_weight(
    class_weight='balanced',
    classes=np.arange(num_classes),
    y=y_train
)
class_weights = {i: float(w) for i, w in enumerate(class_weights_arr)}
print("Class weights:", class_weights)

# --------- SpecAugment (fix Keras 3, aman untuk infer shape) ---------
class SpecAugment(layers.Layer):
    """
    SpecAugment sederhana: 1 masking waktu + 1 masking frekuensi per sample.
    - time_mask: proporsi lebar waktu (0..1), mis. 0.08 = 8%
    - freq_mask: proporsi tinggi frekuensi (0..1), mis. 0.08 = 8%
    Output shape = input shape.
    """
    def __init__(self, time_mask=0.08, freq_mask=0.08, **kwargs):
        super().__init__(**kwargs)
        self.time_mask = float(time_mask)
        self.freq_mask = float(freq_mask)

    def get_config(self):
        cfg = super().get_config()
        cfg.update({"time_mask": self.time_mask, "freq_mask": self.freq_mask})
        return cfg

    def compute_output_shape(self, input_shape):
        return input_shape

    def _mask_one(self, sample):
        # sample: (H, W, C)
        H = tf.shape(sample)[0]
        W = tf.shape(sample)[1]
        C = tf.shape(sample)[2]

        # Lebar band (min 1)
        t_w = tf.maximum(1, tf.cast(tf.round(self.time_mask * tf.cast(W, tf.float32)), tf.int32))
        f_w = tf.maximum(1, tf.cast(tf.round(self.freq_mask * tf.cast(H, tf.float32)), tf.int32))

        # Offset acak
        t0 = tf.cond(W > t_w,
                     lambda: tf.random.uniform([], 0, W - t_w + 1, dtype=tf.int32),
                     lambda: 0)
        f0 = tf.cond(H > f_w,
                     lambda: tf.random.uniform([], 0, H - f_w + 1, dtype=tf.int32),
                     lambda: 0)

        # Bangun mask boolean berbasis range
        cols = tf.range(W)[tf.newaxis, :, tf.newaxis]  # (1, W, 1)
        rows = tf.range(H)[:, tf.newaxis, tf.newaxis]  # (H, 1, 1)

        time_band = tf.logical_and(cols >= t0, cols < t0 + t_w)    # (1, W, 1)
        freq_band = tf.logical_and(rows >= f0, rows < f0 + f_w)    # (H, 1, 1)

        time_mask = tf.cast(tf.logical_not(time_band), sample.dtype)  # 1 di luar band, 0 di dalam
        time_mask = tf.tile(time_mask, [H, 1, C])                     # (H, W, C)

        freq_mask = tf.cast(tf.logical_not(freq_band), sample.dtype)  # 1 di luar band, 0 di dalam
        freq_mask = tf.tile(freq_mask, [1, W, C])                     # (H, W, C)

        # Union of masks → multiply (0 di area yang termask oleh salah satu band)
        mask = time_mask * freq_mask
        return sample * mask

    def call(self, x, training=False):
        # x: (B, H, W, C)
        if not training or (self.time_mask <= 0 and self.freq_mask <= 0):
            return x
        return tf.map_fn(self._mask_one, x)

# --------- Arsitektur CNN ---------
def conv_block(x, filters, k=(3,3), p=(2,2), drop=0.25, l2w=1e-4):
    x = layers.Conv2D(filters, k, padding="same",
                      kernel_regularizer=regularizers.l2(l2w))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Conv2D(filters, k, padding="same",
                      kernel_regularizer=regularizers.l2(l2w))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D(p)(x)
    x = layers.Dropout(drop)(x)
    return x

def build_cnn(input_shape, num_classes):
    inp = layers.Input(shape=input_shape)                 # (H, W, C=3)
    x = SpecAugment(time_mask=0.08, freq_mask=0.08)(inp) # augment ringan saat training

    x = conv_block(x, 32,  k=(3,3), p=(2,2), drop=0.25)
    x = conv_block(x, 64,  k=(3,3), p=(2,2), drop=0.30)
    x = conv_block(x, 128, k=(3,3), p=(2,2), drop=0.35)
    x = conv_block(x, 192, k=(3,3), p=(2,2), drop=0.40)

    x = layers.Conv2D(256, (3,3), padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.5)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)

    out = layers.Dense(num_classes, activation="softmax")(x)
    return models.Model(inp, out)

input_shape = (N_MELS, max_len, X_pad.shape[-1])  # (H, W, C)
model = build_cnn(input_shape, num_classes)
model.summary()

# --------- Compile ---------
opt = optimizers.Adam(learning_rate=1e-3)
model.compile(optimizer=opt,
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

# --------- Callbacks ---------
cbs = [
    callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6, verbose=1),
    callbacks.EarlyStopping(monitor="val_loss", patience=12, restore_best_weights=True, verbose=1),
    callbacks.ModelCheckpoint("best_cnn_melspec.keras", monitor="val_loss", save_best_only=True, verbose=1),
]

# --------- Train ---------
hist = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=40,
    batch_size=32,
    class_weight=class_weights,
    callbacks=cbs,
    verbose=1
)

# --------- Evaluasi ---------
y_val_prob = model.predict(X_val)
y_val_pred = np.argmax(y_val_prob, axis=1)
print("Validation accuracy:", accuracy_score(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred, target_names=class_names, digits=4))
print("Confusion matrix:\n", confusion_matrix(y_val, y_val_pred))

# --------- Plot history ---------
def plot_history(h):
    plt.figure(figsize=(9,6))
    plt.plot(h.history['accuracy'], label='train acc')
    plt.plot(h.history['val_accuracy'], label='val acc')
    plt.title('Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Acc'); plt.grid(True); plt.legend(); plt.show()

    plt.figure(figsize=(9,6))
    plt.plot(h.history['loss'], label='train loss')
    plt.plot(h.history['val_loss'], label='val loss')
    plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.grid(True); plt.legend(); plt.show()

plot_history(hist)
```
